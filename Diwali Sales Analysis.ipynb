#### *Exploratory Data Analysis* ####

### Gender ###

```python
# plotting gender with it's count
ax = sns.countplot(x = 'Gender', data = df, hue = 'Gender')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#plotting a bar chart, gender VS total amount
sales_amount = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_amount, hue = 'Gender')
```

*from the above graph we can see that most of the buyers are females and the purchasimg power of females are greaer than males.*

### Age ###

```python
#plotting age with it's count
ac = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')

for bars in ac.containers:
    ac.bar_label(bars)
```

```python
#plotting bar chart, age VS total amount
amount_age = df.groupby(['Age Group'], as_index = False) ['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = amount_age, hue = 'Amount')
```

*from the above graphs we can see that most of the buyers are from the age group between 26-35 and females.*

### State ###

```python
# plotting bar graphs, state Vs total orders
state_order = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc ={"figure.figsize":(15,5)})
sns.barplot(x = 'State', y ='Orders', data = state_order, hue = 'State')
```

```python
# plotting bar graphs, state Vs total amount
state_amount = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc ={"figure.figsize":(15,5)})
sns.barplot(x = 'State', y ='Amount', data = state_amount, hue = 'Amount')
```

*From the above graphs we can see that most of the orders & total amounts/sales are generated from Uttar Pradesh, Maharashtra, and Karnataka respectively.*

### Marital Status ###

```python
#plotting marital status with it's count
sns.set(rc ={"figure.figsize":(9,5)})
ms = sns.countplot(x='Marital_Status', data = df)

for bars in ms.containers:
    ms.bar_label(bars)
```

```python
# plotting bar chart, marital status VS total sales/amounts
sns.set(rc ={"figure.figsize":(9,5)})
sales_maritalstatus = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x = 'Marital_Status', y = 'Amount', hue = 'Gender', data = sales_maritalstatus)
```

*From the above graphs we can see that most of the buyers are married(women) and they have a high purchasing power.*

### Occupation ###

```python
#plotting occupation with it's count
sns.set(rc ={"figure.figsize":(20,5)})
oc = sns.countplot(x='Occupation', data = df, hue ='Occupation')

for bars in oc.containers:
    oc.bar_label(bars)
```

```python
# plotting bar chart, occupation VS total amount
sns.set(rc ={"figure.figsize":(20,5)})
sales_occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x = 'Occupation', y = 'Amount', hue = 'Occupation', data = sales_occupation)
```

*From the above graphs we can see that most of the buyers are working in IT Sectors, Healthcare, and Aviation sector.*

### Product category ###

```python
# plotting bar graphs, product category Vs total orders
product_order = df.groupby(['Product_Category'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
sns.set(rc ={"figure.figsize":(10,5)})
sns.barplot(x = 'Product_Category', y ='Orders', data = product_order, hue = 'Product_Category')
```

```python
# plotting bar graphs, product category Vs total amount
product_amount = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc ={"figure.figsize":(10,5)})
sns.barplot(x = 'Product_Category', y ='Amount', data = product_amount, hue = 'Product_Category')
```

*From the above graphs we can see that most of the orders and amounts are generated from food, clothing & apparel, and Electronics & gradgts category.*
